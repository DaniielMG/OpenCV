# OpenCV

OpenCV (Open Source Computer Vision Library) es una biblioteca de software de código abierto que proporciona herramientas avanzadas para el desarrollo de aplicaciones en el ámbito de la visión por computadora. Fue desarrollada inicialmente por Intel y lanzada por primera vez en el año 2000. Desde entonces, se ha convertido en una de las bibliotecas más populares y utilizadas en este campo.

### Características Principales

1. **Lenguajes de Programación**: OpenCV proporciona interfaces para varios lenguajes de programación, incluyendo C++, Python y Java.

2. **Plataformas**: Es multiplataforma, lo que significa que se puede ejecutar en diferentes sistemas operativos como Windows, Linux y macOS.

3. **Procesamiento de Imágenes**: Ofrece una gran cantidad de funciones para el procesamiento de imágenes, como manipulación, filtrado, transformaciones, segmentación, extracción de características, entre otros.

4. **Visión por Computadora**: Proporciona herramientas para realizar tareas de visión por computadora tales como reconocimiento de objetos, seguimiento de objetos, reconstrucción 3D, visión estéreo, calibración de cámaras y mucho más.

5. **Aprendizaje Automático**: Incluye módulos para integrar algoritmos de aprendizaje automático, lo que permite implementar y entrenar modelos de machine learning y deep learning directamente desde OpenCV.

6. **Rendimiento**: Está optimizada para un alto rendimiento, con implementaciones eficientes de algoritmos y la capacidad de aprovechar el hardware acelerado.

7. **Interacción con Cámaras**: OpenCV facilita la captura y manipulación de flujos de video de cámaras en tiempo real.

8. **Comunidad y Soporte**: Al ser una biblioteca de código abierto, cuenta con una gran comunidad de desarrolladores y usuarios que contribuyen constantemente al desarrollo y mejora de la biblioteca.

### Aplicaciones de OpenCV

OpenCV se utiliza en una amplia variedad de aplicaciones y sectores, incluyendo:

1. **Robótica**: Para navegación, manipulación y percepción sensorial.
2. **Automoción**: En sistemas de asistencia al conductor y reconocimiento de señales de tráfico.
3. **Seguridad**: En sistemas de vigilancia y reconocimiento facial.
4. **Medicina**: Para análisis de imágenes médicas y diagnóstico.
5. **Realidad Aumentada**: Para crear experiencias interactivas y envolventes.
6. **Fotografía**: Para mejorar imágenes y aplicar efectos especiales.

## Operaciones básicas

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("flor.jpeg")
plt.imshow(img)
```

¡Cuidado! OpenCV utiliza por defecto BGR mientras que Matplotlib usa RGB. Por tanto, para visualizar correctamente las imágenes hay que transformar previamente el espacio de color, BGR->RGB.

```python
img=cv2.imread("flor.jpeg")
 
# Converting BGR color to RGB color format
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
 
#Displaying image using plt.imshow() method
plt.imshow(RGB_img)
```

## Escalado

Métodos de interpolación para cambiar el tamaño de la imagen:

- **cv2.INTER_AREA**: Esto se usa cuando necesitamos reducir una imagen.
- **cv2.INTER_LINEAR**: (Bilineal) Calcula el valor del nuevo pixel como un promedio ponderado de los cuatro píxeles más cercanos.
- **cv2.INTER_CUBIC**: (Bicúbica) Utiliza los 16 píxeles más cercanos (4x4) para calcular el valor del nuevo pixel. Proporciona resultados más suaves en comparación con la interpolación bilineal.
- **cv2.INTER_NEAREST**: Es un método de interpolación vecino más cercano.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("flor.jpeg", 1)
# Loading the image

half = cv2.resize(image, (0, 0), fx = 0.1, fy = 0.1)
bigger = cv2.resize(image, (1050, 1610))

stretch_near = cv2.resize(image, (780, 540),
			interpolation = cv2.INTER_NEAREST)


Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4

for i in range(count):
	plt.subplot(2, 2, i + 1)
	plt.title(Titles[i])
	plt.imshow(images[i])

plt.show()
```

## Recorte

OpenCV proporciona herramientas para realizar una amplia gama de tareas, incluyendo el recorte de imágenes. El recorte de una imagen implica seleccionar una subregión de la imagen y descartar el resto. 

```python
# organizing imports
import cv2
import numpy as np
   
# path to input images are specified and 
# images are loaded with imread command
image = cv2.imread('flor2.jpg')

# color convertion is applied
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
x, y, ancho, alto = 150, 150, 100, 100  # Ejemplo de coordenadas y dimensiones
imagen_recortada = image[y:y+alto, x:x+ancho]

RGB_img = cv2.cvtColor(imagen_recortada, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
# Python program to illustrate
# arithmetic operation of
# addition of two images
   
# organizing imports
import cv2
import numpy as np
   
# path to input images are specified and 
# images are loaded with imread command
image1 = cv2.imread('flor1.jpg')
image2 = cv2.imread('flor2.jpg')
 
# cv2.addWeighted is applied over the
# image inputs with applied parameters
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)
 
# color convertion is applied
RGB_img = cv2.cvtColor(weightedSum, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```
